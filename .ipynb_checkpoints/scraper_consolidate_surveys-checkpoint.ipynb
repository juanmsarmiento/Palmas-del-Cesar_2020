## Preparación de los datos:

```python
# Se importan las encuestas 
xl = pd.ExcelFile('Encuestas.xlsx')
# Se hace un for loop para convertir cada encuesta en un archivo txt independiente
for sheet in xl.sheet_names:
    df = pd.read_excel(xl,sheet_name=sheet)
    df.to_csv(f"{sheet}.txt",index=False, header=False)
# Se crea una lista vacía, para luego llenarla con lo identificadores de cada entrevista
svy_names = []
for sheet in xl.sheet_names:
    svy_names.append(f"{sheet}.txt")
#Se crea 'all_svy.txt', el archivo que contendrá todas las entrevistas una debajo de la otra
with open("all_svy.txt", "w") as outfile:
    for filename in svy_names:
        with open(filename, encoding="utf8") as infile:
            contents = infile.read()
            outfile.write(contents)
```

## Limpieza de datos

```python
# Se hace función con for loop para volver todas las respuesta minúsuclas 
def minuscula(var_list):
    for var in var_list:
        for i in range(len(var)):
            if type(var[i]) == str:
                var[i] = var[i].lower()
    return var_list
```

```python
# Se hace función con for loop para eliminar todas las comas de los datos
def eliminar_comas(var_list):
    for var in var_list:
        for i in range(len(var)):
            if type(var[i]) == str:
                var[i] = var[i].replace(",","")
    return var_list
```

```python
# Se hace función para eliminar las comillas dobles de los datos
def eliminar_comillas(var_list):
    for var in var_list:
        for i in range(len(var)):
            if type(var[i]) == str:
                var[i] = var[i].replace('""', "")
    return var_list
```

```python
# Se hace función para eliminar los espacios antes y después de cada respuesta
def eliminar_espacios(var_list):
    for var in var_list:
        for i in range(len(var)):
            if type(var[i]) == str:
                var[i] = var[i].strip()
    return var_list
```

```python
#Se hace función para eliminar la cadena de caracteres 'na'
def eliminar_na(var_list):
    for var in var_list:
        for i in range(len(var)):
            if type(var[i]) == str:
                var[i] = var[i].replace('na', "")
    return var_list
```

```python
# Se hace función para eliminar los caractéres especiales como tildes 
def eliminar_carespeciales(var_list):
    for var in var_list:
        for i in range(len(var)):
            if type(var[i]) == str:
                var[i] = unidecode.unidecode(var[i])
    return var_list
```

```python
# Se hace función main y se modifican las variables
def main(var_list):
    var_list = minuscula(var_list)
    var_list = eliminar_comas(var_list)
    var_list = eliminar_comillas(var_list)
    var_list = eliminar_espacios(var_list)
    var_list = eliminar_na(var_list)
    var_list = eliminar_carespeciales(var_list)
    return var_list
var_list = main(var_list) 
```

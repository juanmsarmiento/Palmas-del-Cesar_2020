#### Explicación:
- Cada una de estas funciones recupera las respuestas de una pregunta en la encuesta.
- Por esto, cada función ***no*** tiene una descripción, ya que sería redundante escribir que cada una recupera una pregunta.
- Las funciones asumen que todas las encuestas se encuentran en un solo documento (txt), en el cuál buscan las respuestas a través de Regex y generan una lista con las respuestas.
- La extensión de la lista generada en cada función es igual al número de entrevistas, 144.
- El número de cada función indica el orden en el que aparecen las respuestas en la encuesta, a partir de la pregunta de edad. Algunas preguntas (y sus respectivas funciones) fueron saltas para proteger la intimidad de los encuestas, ya que incluían relacionados con el nombre o correo electrónico de los encuestas
